```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sportsreference.ncaab.roster import Roster as ncaa_roster
from sportsreference.ncaab.teams import Teams as ncaa_teams

from sportsreference.nba.roster import Roster as nba_roster
from sportsreference.nba.teams import Teams as nba_teams

from IPython.display import clear_output
```

```python
sns.set()
```

```python
nba_season_sums=pd.DataFrame()
nba_season_sums['Year']=[]
for i in range(2018,1987,-1):
    nba_season_sums=nba_season_sums.append(nba_teams(year=i).dataframes.sum(),ignore_index=True)
    nba_season_sums.loc[2018-i,'Year']=i
    
nba_season_sums['Year']=nba_season_sums['Year'].astype(int)
nba_season_sums=nba_season_sums.drop(columns=['abbreviation','name'])
nba_season_sums=nba_season_sums.set_index('Year')
nba_season_sums
    
```

```python
nba_season_means=pd.DataFrame()
nba_season_means['Year']=[]
for i in range(2018,1987,-1):
    nba_season_means=nba_season_means.append(nba_teams(year=i).dataframes.mean(),ignore_index=True)
    nba_season_means.loc[2018-i,'Year']=i
    
nba_season_means['Year']=nba_season_means['Year'].astype(int)
nba_season_means=nba_season_means.set_index('Year')
nba_season_means
```

```python
nba_shots=nba_season_sums.loc[:,['three_point_field_goal_attempts','two_point_field_goal_attempts']]
nba_shots['3_pct']=nba_shots['three_point_field_goal_attempts']/(nba_shots['three_point_field_goal_attempts']+nba_shots['two_point_field_goal_attempts'])
nba_shots['2_pct']=nba_shots['two_point_field_goal_attempts']/(nba_shots['three_point_field_goal_attempts']+nba_shots['two_point_field_goal_attempts'])

nba_shots
```

```python
plt.bar(nba_shots.index,nba_shots['3_pct'])
line_x=np.array([1988,2018])
line_m,line_b=np.polyfit(nba_shots.index,nba_shots['3_pct'],1)
plt.plot(line_x,line_m*line_x+line_b,color='r')
plt.legend(['Line of Best Fit','NBA 3 Pointers'])
plt.xlabel('Year')
plt.ylabel('Percent of Total Field Goals')
plt.ylim(0,1)
```
